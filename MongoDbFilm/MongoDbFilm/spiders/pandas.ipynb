```python
df_serie = pd.read_csv('top_250_serie.csv')

df_serie
```

```python
df_serie['time'] = df_serie['time'].apply(lambda x : x.split(","))

df_serie
```

```python
df_serie['len'].value_counts()
```

```python
def time_minute (list = list):
    if len(list) == 6 :
        time_minute = str(int(list[1]) * 60 + int(list[4])) + "min"
    elif len(list) == 3 :
        time_minute = str(int(list[1])) + "min"
    else:
        time_minute = "Not time defined"
    return time_minute
```

```python
df_serie['time'] = df_serie['time'].apply(lambda x : time_minute(x))

df_serie.sort_values(by=['score', 'annee'], ascending=False)
```
