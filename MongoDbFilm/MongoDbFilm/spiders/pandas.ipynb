# Nettoyage des fichiers avant transfert dans MongoDB

### Imports

```python
import pandas as pd
```

### Nettoyage fichier top_250_film

```python
def time_minute (list = list):
    if len(list) == 5 :
        time_minute = str(int(list[0]) * 60 + int(list[3]))
    elif len(list) == 2 :
        time_minute = str(int(list[0]))
    else:
        time_minute = 0
    return time_minute
```

```python
df_film = pd.read_csv('csv/top_250_film.csv')

df_film
```

```python
df_film['time'] = df_film['time'].apply(lambda x : x.split(","))
df_film['acteur'] = df_film['acteur'].apply(lambda x : x.split(","))

df_film['time'] = df_film['time'].apply(lambda x : time_minute(x))
# df_film['len'] = df_film['time'].apply(lambda x : len(x))

df_film = df_film.sort_values(by=['score', 'annee'], ascending=False)

df_film
```

```python
df_film['public'].value_counts()
```

```python
df_film['time'].astype(int)
```

```python
df_film.to_csv('csv/top_250_film.csv', index=False)
```

### Nettoyage fichier top_250_serie

```python
df_serie = pd.read_csv('csv/top_250_serie.csv')

df_serie
```

```python
df_serie['saison'].value_counts()
```

```python
def time_minute_serie (list = list):
    if len(list) == 6 :
        time_minute = str(int(list[1]) * 60 + int(list[4]))
    elif len(list) == 3 :
        time_minute = str(int(list[1]))
    else:
        time_minute = 0
    return time_minute
```

```python
df_serie['time'] = df_serie['time'].apply(lambda x : x.split(","))

df_serie['time'] = df_serie['time'].apply(lambda x : time_minute_serie(x))

df_serie = df_serie.sort_values(by=['score', 'annee'], ascending=False)

df_serie
```

```python
df_serie['time'].astype(int)
```

```python
df_serie.info()
```

```python
df_serie.to_csv('csv/top_250_serie.csv', index=False)
```
